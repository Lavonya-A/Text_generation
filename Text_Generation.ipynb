```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
import torch
```

```python
model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
model.eval()
```

```python
input_prompt = "Once upon a time"
input_ids = tokenizer.encode(input_prompt, return_tensors='pt')
```

```python
with torch.no_grad():
    output = model.generate(input_ids, max_length=50, num_return_sequences=1)
```

```python
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Generated Text: \n", generated_text)
```

```python
input_ids = tokenizer.encode(generated_text, return_tensors='pt')
```

```python
with torch.no_grad():
    outputs = model(input_ids, labels = input_ids,  return_dict = True)
    log_likelihood = outputs.loss
```

```python
perplexity = torch.exp(log_likelihood).item()
print(f'Perplexity of the generated text : {perplexity:.4f}')
```
